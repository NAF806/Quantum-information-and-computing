## *Superposition* #

*As mentioned before in the introduction, the Hadammard gate is used to create superposition and can be explicity shown as such:*

\begin{equation}
H = \frac{1}{\sqrt 2} \begin{bmatrix} 1 & 1\\ 1 & -1 \end{bmatrix}
\end{equation}

*We can then see how this would act on the $|0\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix} $:*

\begin{equation}
H |0\rangle = \frac{1}{\sqrt 2} \begin{bmatrix} 1 & 1\\ 1 & -1 \end{bmatrix} \begin{bmatrix} 1 \\ 0 \end{bmatrix} = \frac{1}{\sqrt 2} \begin{bmatrix} 1 \\1 \end{bmatrix} = \frac{1}{\sqrt 2} (|0 \rangle + | 1 \rangle)
\end{equation}

*The output from this an equal superposition of states between $|0 \rangle$ and $|1 \rangle$ with an amplitude of $ \frac{1}{\sqrt 2} $ for each state. This means the probability of measuring state, which is given by the amplitude squared, is $ |\frac{1}{\sqrt 2}|^2 = \frac 12 $ This can be expanded to multi-qubit systems by applying a Hadammard gate to each individual qubit and then computing the tensor product of the qubits in the system.*


## *Proof for no-cloning theorem* ## 

*Consider the unkown quantum state $|\psi \rangle$:*

\begin{equation}
|\psi \rangle \alpha |0 \rangle +\beta |1 \rangle 
\end{equation}

*where $ \alpha $ and $ \beta $ are complex numbers that satisfy the normalization condition.*

*To clone $|\psi \rangle $, we start by computing a tensor product with a $|0 \rangle $, so that the intial state becomes:*

\begin{equation}
|\psi \rangle \otimes |0 \rangle = ( \alpha |0 \rangle + \beta | 1 \rangle ) \otimes |0 \rangle  = \alpha | 0 \rangle \otimes |0 \rangle + \beta | 1 \rangle \otimes | 1 \rangle 
\end{equation}

*The goal of cloning is to apply such a unitary operation U that it transforms $|\psi \rangle \otimes |0\rangle $ into $|\psi \rangle \otimes |\psi \rangle $, we want the result to be:* 

\begin{equation}
|\psi \rangle \otimes |\psi \rangle = ( \alpha | 0 \rangle + \beta | 1 \rangle ) \otimes ( \alpha |0 \rangle + \beta | 1 \rangle)
\end{equation}

*If cloning were possible, the unitary operator U would have to satisfy:*

\begin{equation}

U(|\psi \rangle \otimes |0 \rangle) = |\psi \rangle \otimes \psi \rangle 
\end{equation}

\begin{equation}

U(|\phi \rangle \otimes |0 \rangle) = |\phi \rangle \otimes \phi \rangle 

\end{equation}

*By linearity of quantum mechanics, the action U on a superposition of $ |\psi \rangle $ and $|\phi \rangle $ should be:*

\begin{equation}
U ((|\psi \rangle + |\phi \rangle) \otimes |0\rangle) = U(|\psi \rangle \otimes |0\rangle) + U(|\psi \rangle \otimes |0\rangle)
\end{equation}

*If U were a cloning operator, this would imply:*

\begin{equation}
U ((|\psi \rangle + |\phi \rangle) \otimes |0\rangle) = (|\psi \rangle + |\phi \rangle ) \otimes (|\psi \rangle + |\phi \rangle )
\end{equation}

*Evaluating both the above equations:*
 * *Due to linearity:*
 \begin{equation}
 U(|\psi \rangle \otimes |0\rangle) + U(|\psi \rangle \otimes |0\rangle) = |\psi \rangle \otimes |\psi \rangle + |\phi \rangle \otimes |\phi \rangle
 \end{equation}

 * expanding ($(|\psi \rangle + |\phi \rangle ) \otimes (|\psi \rangle + |\phi \rangle$)

 \begin{equation}
 (|\psi \rangle + |\phi \rangle ) \otimes (|\psi \rangle + |\phi \rangle ) = |\psi \rangle \otimes |\psi \rangle + |\phi \rangle \otimes |\phi \rangle + |\psi \rangle \otimes |\phi \rangle + |\phi \rangle \otimes |\psi \rangle
 \end{equation}

*It is clear that these 2 cannot be equal as the latter equation consists of cross terms that the equation from linearity lacks. Since no linear (unitary) operator can satisfy both these conditions simultaneously for an arbitary $|\psi \rangle $ and $|\phi \rangle$ , it is impossible to construt a unitary operator that clones quantum states. This proves the no-cloning theorem.*
